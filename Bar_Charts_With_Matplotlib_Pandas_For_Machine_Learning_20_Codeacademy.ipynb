```python
#Importieren der Wergzeuge, falls von anfang an bei JupyterNB oder VC oder Pyton IDL etc. pip install nicht vergessen
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Bar Plot
#my_df.plot(kind = 'bar') #zeigt nich die absolute Funktion (nicht nur absolute values)
my_df.abs().plot(kind = 'bar')
```

```python
# bar plot Stacked
my_df.abs().plot(kind= "bar", stacked = True)
```
